```python
#parameters optimization for Knn model in smote dataset
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import f1_score

knn_classifier = KNeighborsClassifier()

param_grid = {
    'n_neighbors': [3, 5, 7, 9],  
    'weights': ['uniform', 'distance'],  
    'p': [1, 2], 
}

best_params = {}
best_f1_score = 0

for n_neighbors in param_grid['n_neighbors']:
    for weights in param_grid['weights']:
        for p in param_grid['p']:
            print("\n for n_neighbors=" + str(n_neighbors) + ", weights=" + weights + ", p=" + str(p) + ".....")
            knn_classifier.set_params(n_neighbors=n_neighbors, weights=weights, p=p)
            grid_search = GridSearchCV(knn_classifier, param_grid={}, cv=5, scoring='accuracy')
            grid_search.fit(scaled_train_smote, y_train_smote)
            val_pred = grid_search.predict(scaled_val_smote)
            avg_f1_score_knn = f1_score(val_test_smote, val_pred, average='macro')
            if avg_f1_score_knn > best_f1_score:
                best_f1_score = avg_f1_score_knn
                best_params = {'n_neighbors': n_neighbors, 'weights': weights, 'p': p}

print('\n The optimal parameters for this data are: ' + str(best_params))
```

```python
from sklearn import metrics
import torch
import torch.nn as nn
import torch.optim as optim



# Convert pandas DataFrames to PyTorch tensors
X_train_tensor = torch.tensor(scaled_train_smote, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train_smote.values - 1, dtype=torch.long)  # Adjust labels to start from 0
X_test_tensor = torch.tensor(scaled_test_smote, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test_smote.values - 1, dtype=torch.long)  # Adjust labels to start from 0
X_val_tensor = torch.tensor(scaled_val_smote, dtype=torch.float32)
y_val_tensor = torch.tensor(val_test_smote.values - 1, dtype=torch.long)  # Adjust labels to start from 0

total_loss=[]
# Step 2: Define the MLP model architecture
class MLP(nn.Module):
    def __init__(self, input_size, hidden_size1, hidden_size2,hidden_size3,hidden_size4,output_size):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size1)
        self.fc2 = nn.Linear(hidden_size1, hidden_size2)
        self.fc3 = nn.Linear(hidden_size2, hidden_size3)
        self.fc4 = nn.Linear(hidden_size3, hidden_size4)
        self.fc5 = nn.Linear(hidden_size4, output_size)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        x = self.relu(x)
        x = self.fc4(x)
        x = self.relu(x)
        x = self.fc5(x)
        return x


# Step 3: Initialize the model and define the loss function and optimizer
input_size = new_train_smote.shape[1]
hidden_size = 12
hidden_size2=8
hidden_size3=12
hidden_size4=6
output_size = len(set(y_train_smote))

model = MLP(input_size, hidden_size,hidden_size2,hidden_size3,hidden_size4,output_size)
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.1)

# Step 4: Train the model
num_epochs = 100
batch_size = 32

for epoch in range(num_epochs):
    # Perform forward pass
    outputs = model(X_train_tensor)
    loss = loss_fn(outputs, y_train_tensor)
    

    # Perform backward pass and optimize
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Print training progress
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item()}")

# Step 5: Evaluate the model
with torch.no_grad():
    model.eval()
    val_outputs = model(X_val_tensor)
    val_loss = loss_fn(val_outputs, y_val_tensor)
    val_predictions = torch.argmax(val_outputs, dim=1)
    total_loss.append([loss.item(),val_loss.item()])
    accuracy = (val_predictions == y_val_tensor).sum().item() / len(y_val_tensor)
    print(f"Validation Loss: {val_loss.item()}, Accuracy: {accuracy}")

# Step 6: Make predictions on the test set
test_outputs = model(X_test_tensor)
test_predictions = torch.argmax(test_outputs, dim=1)
# Process the predictions as needed

test_outputs = model(X_test_tensor)
test_predictions = torch.argmax(test_outputs, dim=1)




def plot_confusion_matrix(labels, pred_labels):

    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(1, 1, 1)
    cm = metrics.confusion_matrix(labels, pred_labels)
    cm = metrics.ConfusionMatrixDisplay(cm, display_labels=range(0,7))
    cm.plot(values_format='d', cmap='Blues', ax=ax)

plot_confusion_matrix(y_test_tensor, test_predictions)

```
